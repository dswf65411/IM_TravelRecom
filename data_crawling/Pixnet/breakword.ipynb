```python
def replace_dot(strin):
    '''
        input  :str
        output :str with "," "，" "、" "。" "！" "!" "？" "?"... replece by " ", and 全形字 replaced by 半形字
    '''
    replaceDot = set([",","，","、","。","！","!","？","?",":","：",";","；","「","」","【","】","（","）","『","』","．",
                     "「","」","【","】","『","』",'‘', '’', '“', '”', '〝', '〞', '‵', '′', '〃','^','~','.', 
                      '～', '＄', '％', '＠', '＆', '＃', '＊', '‧', '；', '︰','…', '‥', '﹐', '﹒', '˙', '· ﹔',
                      '﹕', '‘', '’', '“', '”', '〝', '〞', '‵', '′', '〃', '_','=','＝','+','-','(',')',' ']
                    )
#   replaceDot = set(["「","」","【","】","『","』",'‘', '’', '“', '”', '〝', '〞', '‵', '′', '〃'])
    replaceDot2 = set(['，', '、', '。', '．', '？', '！', '～', '＄', '％', '＠', '＆', '＃', '＊', '‧', '；', '︰','…', '‥', '﹐', '﹒', '˙', '· ﹔', '﹕', '‘', '’', '“', '”', '〝', '〞', '‵', '′', '〃'])
    replaceBig = "ＡＢＣＤＥＦＧＨＩＪＫＬＭＮＯＰＱＲＳＴＵＶＷＸＹＺａｂｃｄｅｆｇｈｉｊｋｌｍｎｏｐｑｒｓｔｕｖｗｘｙｚ１２３４５６７８９０"
    replaceSmall="ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz1234567890"
    for dotTmp in replaceDot:
#         strin = strin.replace(dotTmp," ")
        strin = strin.replace(dotTmp,"")
    for i in range(len(replaceBig)):
        strin = strin.replace(replaceBig[i],replaceSmall[i])
    return strin
```

```python
uri = "mongodb://travelteam:okdavid@ds011432.mlab.com:11432/imtravel" 
client = MongoClient(uri)
db = client['imtravel']
collection = db['台南']
```

```python
post = list(collection.find({},{"site":1,"_id":0}))
sites = []
for item in post:
    sites.append(item['site'])
```

```python
doc = list(collection.find({},{'comment':1,'_id':0,'site':1}))
```

```python
comments = {}
for obj in doc:
    obj['comment'][0]['content'].extend(obj['comment'][1]['content'])
    comments[obj['site']] = obj['comment'][0]['content']
```

```python
print(comments['五妃廟'])
```

```python
for site in sites:
    for i in range(0,len(comments[site])):
        content = comments[site][i][0]+comments[site][i][1]
        comments[site][i] = replace_dot(content)
    comments[site] = "".join(comments[site])
```

```python
result = {}
for site in sites:
    temp = []
    for i in range(0,len(comments[site])-1):
        temp.append(comments[site][i]+comments[site][i+1])
    result[site] = temp
```
